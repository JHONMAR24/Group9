```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/akmand/datasets/master/openintro/kobe.csv')
df
```

```python
def createDB(DBname, FileName, TableName):
    conn = sqlite3.connect(DBname) 
    cur = conn.cursor()

    df = pd.read_csv(FileName, nrows=200)

    df.to_sql(TableName, conn, if_exists="replace", index=False) 
    
    result = pd.read_sql_query(f"SELECT * FROM {TableName}", conn)
    
    conn.commit()
    conn.close()
    
    return result
```
