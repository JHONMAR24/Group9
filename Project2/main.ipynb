```python
#Libraries required to operate the code:
import pandas as pd
import sqlite3
```

```python
#Data Collection stage: This information downloaded is stored into 'df' variable
df = pd.read_csv('https://raw.githubusercontent.com/akmand/datasets/master/openintro/kobe.csv')
df
```

```python
#Function CreateDB: We pass three variables:
#      -DBname: This value names new DB file that will be created. 
#      -Filename: It is the name of the file/location that contains the data frame.
#      -TableName: Name of the table where the data will be stored.
def createDB(DBname, FileName, TableName):
    conn = sqlite3.connect(DBname) 
    cur = conn.cursor()

    df = pd.read_csv(FileName, nrows=100)

    df.to_sql(TableName, conn, if_exists="replace", index=False) 
    
    result = pd.read_sql_query(f"SELECT * FROM {TableName}", conn)
    
    conn.commit()
    conn.close()
    
    return result
```

```python
#Use of the createDB function with their input variables:

createDB("kobe.db",'https://raw.githubusercontent.com/akmand/datasets/master/openintro/kobe.csv',"kobe_table")
```
